# Variables

Variables are containers for storing data values. A variable is created the moment you first assign a value to it.

```python
num = 10
print(num)
```

```python
name = "Abdullah"
print(name)
```

Variables do not need to be declared with any particular type, and can even change type after they have been set.

```python
# x here is of type int
x = 20

# x now is of type str
x = "Khan"
print(x)
```

## Casting

If you want to specify the data type of a variable, this can be done with casting.

```python
x = str(10)
y= int(10)
z = float(10)

print(x)
print(y)
print(z)
x,y,z
```

## Data Type

We can get the data type of variable using the type() function.

```python
type(num)
```

```python
type(name)
```

## Case Sensitive

In Python variable name are case sensitive.

```python
b = 5
B = "khan"

#B will not overwrite b.
print(b)
print(B)
```

## Variable Name

A variable can have a short name like x and y or a more descriptive name like age, carname, total_volume

### Legal Variable Names

```python
myname = "Abdullah"
my_name = "Abdullah"
_my_name = "Abdullah"
myName = "Abdullah"
myname2 = "Abdullah"

print(myname)
print(my_name)
print(_my_name)
print(myName)
print(myname2)
```

### Illegal Names

```python
2myname = "Abdullah"
my-name = "Abdullah"
my name = "Abdullah"

print(2myname)
print(my-name)
print(my name)
```

## Multi words Variables

Variable names with more than one word can be difficult to read.
There are several techniques you can use to make them more readable:

### Camal Case Variable
Each word, except the first, starts with a capital letter:

```python
myFullName = "Abdullah Khan"
print(myFullName)
```

### Pascal case variable
Each word starts with a capital letter:

```python
MyFullName = "Abdullah Khan"
print(MyFullName)
MyFullName
```

### Snake case

```python
my_full_name = "Abdullah khan"
print(my_full_name)
```

## Many Values to Multiple Variables

Python allows you to assign values to multiple variables in one line:

```python
x,y,z = "Eagle", "Parrot", "Pigeon"

print(x)
print(y)
print(z)
```

## One value to Multiple Variables.
We can assign the same value to multiple variables in one line

```python
m = n = o = "Eagle"

print(m)
print(n)
print(o)
```

## Unpack A Collection
If you have a collection of values in a list, tuple etc. Python allows you to extract the values into variables. This is called unpacking.

```python
birds = [ "Eagle", "Parrot", "Pigeon"]
x1, y1, z1 = birds

print(x1)
print(y1)
print(z1)
```

## Output Variables
The print() function is often used to output variables.

```python
name = "My name is Abdullah"
print(name)
```

In the print() function, you output multiple variables, separated by a comma:

```python
a = "My"
b = "name"
c = "is"
d = "Abdullah"

print(a,b,c,d)
```

We can also use the + operator to output multiple variables:

```python
a = "My "
b = "name "
c = "is "
d = "Abdullah"

print(a + b + c + d)
```

For numbers, the + character works as a mathematical operator:

```python
num1 = 30
num2 = 20
num3 = 50

print(num1 + num2 + num3)
```

In the print() function, when you try to combine a string and a number with the + operator, Python will give you an error:

```python
num = 10
str = "khan"

print(num + str)
```

The best way is to separate the variable in print() function with comma.

```python
num = 10
str = "khan"

print(num, str)
```

## Global Variable

Variables that are created outside of a function are known as global variables. 
Global variables can be used by everyone, both inside of functions and outside.

```python
name = "Abdullah"

def my_name():
    print("My name is " + name)

my_name()
```

## Local Variable
If you create a variable with the same name inside a function, this variable will be local, and can only be used inside the function

```python
# This is Global variable
name = "Afridi"

def my_sir_name():
    # This is Local variable
    name = "Khan"
    print("My name is ", name)
    
my_sir_name()

print("My name is ", name)
```
