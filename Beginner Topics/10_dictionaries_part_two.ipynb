# Add Items
Adding an item to the dictionary is done by using a new index key and assigning a value to it:


```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}

my_dict["color"] = "White"

print(my_dict)
```

# Remove items
There are several methods to remove items from a dictionary:

### pop()
The pop() method removes the item with the specified key name:


```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_dict.pop("model")

print(my_dict)
```

### popitem()
The popitem() method removes the last inserted item (in versions before 3.7, a random item is removed instead):


```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_dict.popitem()

print(my_dict)
```

### del keyword
The del keyword removes the item with the specified key name, Also it can delete the dictionarie completely:


```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
del my_dict["year"]

print(my_dict)
```

### clear()
The clear() method empties the dictionary:

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_dict.clear()

print(my_dict)
```

# Loop Dictionaries

### for loop
We can loop through a dictionary by using a for loop.

When looping through a dictionary, the return value are the keys of the dictionary, but there are methods to return the values as well.

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
for x in my_dict:
    print(x)
```

```python
# Print all values in the dictionary, one by one:
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
for x in my_dict:
    print(my_dict[x])
```

```python
# You can also use the values() method to return values of a dictionary: same keys() method for keys.
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
for x in my_dict.values():
    print(x)
```

```python
# Loop through both keys and values, by using the items() method:
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
for x,y in my_dict.items():
    print(x,y)

```

# Copy Dictionaries
There are ways to make a copy, one way is to use the built-in Dictionary method copy().


```python
#Make a copy of a dictionary with the copy() method:
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
new_dict = my_dict.copy()
print(new_dict)
```

```python
# Another way to make a copy is to use the built-in function dict().
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
new_dict = dict(my_dict)
print(new_dict)
```

# Nested Dictionaries
A dictionary can contain dictionaries, this is called nested dictionaries.


```python
# Create a dictionary that contain three dictionaries:
my_family = {
     "child1": {
         "name": "Ali",
         "age": 15
     },
     "child2": {
         "name": "Salman",
         "age": 13
     },
     "child3": {
         "name": "Khan",
         "age": 10
     }
 }

print(my_family)
```

### Access Items
To access items from a nested dictionary, you use the name of the dictionaries, starting with the outer dictionary:

```python
my_family = {
     "child1": {
         "name": "Ali",
         "age": 15
     },
     "child2": {
         "name": "Salman",
         "age": 13
     },
     "child3": {
         "name": "Khan",
         "age": 10
     }
 }

print(my_family["child1"]["name"])
print(my_family["child1"]["age"])
print(my_family["child3"]["name"])
print(my_family["child3"]["age"])
```
