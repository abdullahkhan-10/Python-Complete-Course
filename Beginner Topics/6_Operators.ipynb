# Python Operators
Operators are used to perform operations on variables and values

## Arithmetic Operators
Arithmetic operators are used with numeric values to perform common mathematical operations

```python
x = 35
y = 4

# Addition
print(x + y)

#Subtraction
print(x - y)

#Multiplication
print(x * y)
```

```python
#Division (Return float value)
print(x / y)
```

```python
# Modulo (Return remainder of division operation)
print(x % y)
```

```python
# Floor division always returns an integer, It rounds DOWN to the nearest integer
print(x // y)
```

```python
# Exponentiation (Power Operator), an arithmetic operator, it raises the first number to the power of the second
print(x ** y)
```

# Assignment Operators
Assignment operators are used to assign values to variables

```python
x = 10

print(x)
```

```python
x = 5

# same as x = x + 3
x += 3

print(x)
```

```python
x = 6

# same as x = x - 2
x -= 2

print(x)
```

```python
x = 4

# same as x = x * 3
x *= 3

print(x)
```

```python
x = 20

# same as x = x / 3
x /= 4

print(x)
```

# Comparison Operators
Comparison operators are used to compare two values

```python
x = 15
y = 12

# Equal
print(x == y)

# Not Equal
print(x != y)

# Greater Than
print(x > y)

# less than
print(x < y)

# Greater than or Equal
print(x >= y)

# Less than or equal
print(x <= y)
```

# Logical Operators
Logical operators are used to combine conditional statements

### and operator
Return true if both statement are true.

```python
x = 10

print(x > 0 and x < 20)
```

### or operator
Returns True if one of the statements is true

```python
x = 5

print(x < 6 or x > 10)
print(x < 4 or x > 10)
```

### not operator
Reverse the result, returns False if the result is true

```python
x = 5

# Here result should be True, but not operator make it false
print(not(x == 5))

# Here result should be False, but not operator make it True.
print(not(x != 5))
```

# Identity Operators
Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.

### is operator
Returns True if both variables are the same object

```python
x = ["Eagle", "Parrot"]
y = ["Eagle", "Parrot"]
z = x

# is - Checks if both variables point to the same object in memory
# z and x both point to first objects in memory.
print(x is z)

# Return False, because both objects have different places in memory
print(x is y)

# ==  Checks if the values of both variables are equal
print(x == y)
```

### is not operator
The is not operator returns True if both variables do not point to the same object

```python
x = ["Eagle", "Parrot"]
y = ["Eagle", "Parrot"]

print(x is not y)
```

# Membership Operators
Membership operators are used to test if a sequence is presented in an object

### in operator
Returns True if a specified value is present in the object

```python
birds = ["Eagle", "Parrot", "Pigeon"]

print("Parrot" in birds)
print("Falcon" in birds)
```

### not in operator
Returns True if a specified value is not present in the object

```python
birds = ["Eagle", "Parrot", "Pigeon"]

print("Sparrow" not in birds)
print("Eagle" not in birds)
```

## Membership operator in string
Membership operator also works in string

```python
name = "Abdullah Khan"

print("A" in name)
print("abdullah" in name)
print("afridi" not in name)
```

# Bitwise Operators
Bitwise operators are used to compare (binary) numbers

### & operator (name AND)
The & operator compares each bit and set it to 1 if both are 1, otherwise it is set to 0

```python
# The binary representation of 6 is 0110
# The binary representation of 3 is 0011

print( 6 & 3)

#Then the & operator compares the bits and returns 0010, which is 2 in decimal.
```

### | operator (name OR)
The | operator compares each bit and set it to 1 if one or both is 1, otherwise it is set to 0

```python
# The binary representation of 6 is 0110
# The binary representation of 3 is 0011

print( 6 | 3)
#Then the | operator compares the bits and returns 0111, which is 7 in decimal.
```
