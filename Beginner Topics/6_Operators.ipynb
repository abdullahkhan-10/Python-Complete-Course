# Python Operators
Operators are used to perform operations on variables and values

## Arithmetic Operators
Arithmetic operators are used with numeric values to perform common mathematical operations

```python
x = 35
y = 4

# Addition
print(x + y)

#Subtraction
print(x - y)

#Multiplication
print(x * y)
```

```python
#Division (Return float value)
print(x / y)
```

```python
# Modulo (Return remainder of division operation)
print(x % y)
```

```python
# Floor division always returns an integer, It rounds DOWN to the nearest integer
print(x // y)
```

```python
# Exponentiation (Power Operator), an arithmetic operator, it raises the first number to the power of the second
print(x ** y)
```
