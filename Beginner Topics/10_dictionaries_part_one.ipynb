# Python Dictionaries
Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

```python
# Note As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.
```

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}

print(my_dict)
```

### Dictionary Items
Dictionary items are presented in key:value pairs, and can be referred to by using the key name.

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}

print(my_dict["model"])
```

### No Duplicates
Dictionaries cannot have two items with the same key:

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016,
    "year": 2019
}

print(my_dict)
```

### Dictionary Length
To determine how many items a dictionary has, use the len() function:

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}

print(len(my_dict))
```

### Items Data Types
The values in dictionary items can be of any data type:

```python
my_dict = {
    "brand": "Gulfstream",
    "speed": True,
    "year": 2016,
    "color": ["Black", "White", "Golden"]
}

print(my_dict)
```

### type()
From Python's perspective, dictionaries are defined as objects with the data type 'dict':

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}

print(type(my_dict))
```

### dict() Constructor
It is also possible to use the dict() constructor to make a dictionary.

```python
my_dict = dict(brand = "Gulfstream", model = "G600", year = 2016)

print(my_dict)
```

# Access Dictionary Items
You can access the items of a dictionary by referring to its key name, inside square brackets:

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_item = my_dict["brand"]
print(my_item)
```

```python
# There is also a method called get() that will give you the same result:
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_item = my_dict.get("brand")
print(my_item)
```

### Get Keys
The keys() method will return a list of all the keys in the dictionary.

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_keys = my_dict.keys()
print(my_keys)
```

```python
# Note. The list of the keys is a view of the dictionary,
# meaning that any changes done to the dictionary will be reflected in the keys list.
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_keys = my_dict.keys()
# Before adding new item
print(my_keys)

my_dict["color"] = "White"
# After adding new item
print(my_keys)
```

### Get Values
The values() method will return a list of all the values in the dictionary.

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_values = my_dict.values()
print(my_values)
```

```python
# Add a new item to the original dictionary, and see that the values list gets updated as well:
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_values = my_dict.values()
# Before adding new item
print(my_values)

my_dict["newModel"] = "G800"
# After adding new item
print(my_values)
```

### Get Items
The items() method will return each item in a dictionary, as tuples in a list.

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_items = my_dict.items()

print(my_items)
```

```python
# Make a change in the original dictionary, and see that the items list gets updated as well:
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
my_items = my_dict.items()
# Before the change of item
print(my_items)

my_dict["year"] = 2019
# After the change of item
print(my_items)
```

### if Key Exists
To determine if a specified key is present in a dictionary use the in keyword:

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}

if "model" in my_dict:
    print("Yes, model is one of the key available in dictionary")
```

# Change Dictionary Items

### Change Values
We can change the value of a specific item by referring to its key name:

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}
# Change the model to G800
my_dict["model"] = "G800"

print(my_dict)
```

### Update Dictionary
The update() method will update the dictionary with the items from the given argument.

The argument must be a dictionary, or an iterable object with key:value pairs.

```python
my_dict = {
    "brand": "Gulfstream",
    "model": "G600",
    "year": 2016
}

my_dict.update({"year": 2024})
print(my_dict)
```
