# Strings

Strings in python are surrounded by either single quotation marks, or double quotation marks.

'Abdullah' is the same as "Abdullah".

```python
print("I am 'Abdullah khan'")
print("it's my full name")
print('I like reading "books"')
```

### Multi Lines string
We can assign a multiline string to a variable by using three quotes:

```python
name = "My name is Abdullah khan Afridi, I am from District Khyber, Pakistan"
print(name)
```

## String Are Array
Like many other popular programming languages, strings in Python are arrays of unicode characters.

```python
name = "Abdullah khan"
# we use square bracket to access elements of strings.
print(name[2])
print(name[5])
print(name[9])
print(name[3:8])
```

### loop through string

```python
for element in "Khan":
     print(element)
 
```

### String length

```python
name = "Abdullah khan"
print(len(name))
```

# String Slicing
Slicing is used to return range of characters of a string.

```python
name = "Hello, world"
print(name[2:5])
```

```python
# Slice from start to index 5
name = "Hello, world"
print(name[:5])
```

```python
# Slice from index 2 to end
name = "Hello, world"
print(name[2:])
```

```python
# Use negative indexes to start the slice from the end of the string.
name = "Hello, world"
print(name[-4:-1])
```

```python
name = "Hello, world"
print(name[-9:-2])
```

# String Modification
Python has a set of built-in methods that you can use to modify strings

```python
# UPPER CASE
name = "abdullah khan"
print(name.upper())
```

```python
# lower case
name = "ABDULLAH KHAN"
print(name.lower())
```

```python
# Remove white spaces before and after string 
name = " Abdullah khan   "
print(name.strip())
```

```python
# The replace() method is used to replace character of string with another characters.
name = "Abdullah khan"
print(name.replace("A", "E"))
```

```python
# The split() method returns a list where the text between the specified separator becomes the list items
name = "Abdullah, khan"
print(name.split(","))
```

# String Concatenation

```python
# To concatenate, or combine, two strings you can use the + operator
first_name = "Abdullah "
last_name = "khan"
full_name = first_name + last_name
print(full_name)
```

## F-String
F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.

To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations.

```python
age = 23
name = f"My name Abdullah, I am {age} years old."
print(name)
```

```python
# A placeholder can contain Python code, like math operations
value = f"The price of new book is {10 + 15} dollars."
print(value)
```

# Escape character 
To insert characters that are illegal in a string, use an escape character.
An escape character is a backslash \ followed by the character you want to insert.
An example of an illegal character is a double quote inside a string that is surrounded by double quotes:

```python
txt = "We are the so-called "Vikings" from the north."
print(txt)
```

```python
# To fixt this use escape character.
txt = "We are the so-called \"Vikings\" from the north."
print(txt)
```
