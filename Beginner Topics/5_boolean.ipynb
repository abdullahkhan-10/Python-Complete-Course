# Boolean
Booleans represent one of two values: True or False.
We can evaluate any expression in Python, and get one of two answers, True or False.

```python
print(10 > 9)
print(10 == 9)
```

```python
x = 75
y = 20
if y > x:
    print("y is greater than x")
else:
    print("y is not greater than x")
          
```

## Evaluate values and variables
Almost any value is evaluated to True if it has some sort of content

```python
#Any string is True, except empty strings
print(bool("Khan"))

#Any number is True, except 0.
print(bool(10))

#Any list, tuple, set, and dictionary are True, except empty ones. 
birds = 
print(bool(birds))

print(bool(["Eagle", "Parrot", "Pigeon"])
```

Empty values, such as (), [], {}, "", the number 0, and the value None. And of course the value False evaluates to False.

```python
#The False evaluate to false.
print(bool(False))

#The None Evaluate to false
print(bool(None))

#The Zero Evaluate to false
print(bool(0))

#The empty set, Dic, list and tuple Evaluate to false
print(bool({}))
print(bool([]))
print(bool(()))
```

## Functions can Return a Boolean
We can create functions that returns a Boolean Value

```python
def bool_func():
    return True
    
print(bool_func())
```

```python
def bool_func1():
    return False
    
print(bool_func1())
```

We can execute code based on the Boolean answer of a function

```python
def my_func():
    return True
    
if my_func():
    print("Yes!")
else:
    print("No!")
```

```python
def my_func2():
    return False
    
if my_func2():
    print("Yes!")
else:
    print("No!")
```
