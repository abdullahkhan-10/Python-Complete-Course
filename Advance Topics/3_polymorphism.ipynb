# Python Polymorphism
The word "polymorphism" means "many forms", and in programming it refers to methods/functions/operators with the same name that can be executed on many objects or classes.

An example of a Python function that can be used on different objects is the len() function.

### String
For strings len() returns the number of characters:


```python
greeting = "Assalam o alaikum"

print(len(greeting))
```

### tuple
For tuples len() returns the number of items in the tuple:


```python
birds_tuple = ("Eagle", "Pigeon", "Falcon", "Parrot")

print(len(birds_tuple))
```

### Dictionary
For dictionaries len() returns the number of key/value pairs in the dictionary:


```python
my_jet = {
    "brand" : "Gulfstream",
    "model" : "G800",
    "year" : 2025
}

print(len(my_jet))
```

# Class Polymorphism
Polymorphism is often used in Class methods, where we can have multiple classes with the same method name.

For example, say we have three classes: Car, Boat, and Plane, and they all have a method called move()

```python
class Car:
    def __init__(self, brand, model):
        self.my_brand = brand
        self.my_model = model

    def move(self):
        print("Drive!")
        
class Boat:
    def __init__(self, brand, model):
        self.my_brand = brand
        self.my_model = model

    def move(self):
        print("Sail!")

class Plane:
    def __init__(self, brand, model):
        self.my_brand = brand
        self.my_model = model

    def move(self):
        print("Fly!")

car1 = Car("Toyota", "Lexus")
boat1 = Boat("Brunswick", "VR6 Bowrider")
plane1 = Plane("Boeing", "747")

for x in (car1, boat1, plane1):
    print(x.my_brand)
    x.move()

# Because of polymorphism we can execute the same method for all three classes.
```

## Inheritance Class Polymorphism
When we make a parent class called Vehicle, and make Car, child classes of Vehicle. Child classes inherits the properties and methods from the parent class.

```python
class Vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("The Toyota Lexus move fast")

class Car(Vehicle):
  pass

car1 = Car("Toyota", "Lexus")
car1.move()

# In the example above you can see that the Car class is empty, but it inherits brand, model, and move() from Vehicle.
```

```python
# The Plane class also inherit brand, model, and move() from Vehicle, but they both override the move() method.
```

```python
class Vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
      print("The vehicle move")

class Plane(Vehicle):
  def move(self):
    print("The plane Fly")

plane1 = Plane("Boeing", "747")
plane1.move()

# The move() method of Plane class override the move() method of Vehicle. 
```
