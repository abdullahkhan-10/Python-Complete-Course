# File Handling
File handling is an important part of any web application.

Python has several functions for creating, reading, updating, and deleting files

### Open file
The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode

There are four different methods (modes) for opening a file:

"r" - Read - Default value. Opens a file for reading, error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns an error if the file exists

In addition you can specify if the file should be handled as binary or text mode

"t" - Text - Default value. Text mode

"b" - Binary - Binary mode (e.g. images)


### Open and Read file

```python
my_file = open("demo.txt", "r")

print(my_file.read())

# The open() function returns a file object, which has a read() method for reading the content of the file:
```

### Read Lines
We can return one or many line by using the readline() and readlines() methods.

```python
# The readline() method is used to read the first line of the demo.txt file.
my_file = open("demo.txt", "r")

print(my_file.readline())
```

```python
# The readlines() method is used to read the data in list formate.
my_file = open("demo.txt", "r")

print(my_file.readlines())
```

### with statement
To opne a file we can use with statement.

```python
with open("demo.txt", "r") as my_file:
    print(my_file.read())
```

### Close Files
It is a good practice to always close the file when you are done with it.

```python
my_file = open("demo.txt", "r")

print(my_file.read())
my_file.close()

# Note: You should always close your files. In some cases, due to buffering, changes made to a file may not show until you close the file.
```

### Read specific Parts
By default the read() method returns the whole text, but we can also specify how many characters we want to return:


```python
my_file = open("demo.txt", "r")

print(my_file.read(10))
my_file.close()
```

# Write File 
To write to an existing file, you must add a parameter to the open() function:

"a" - Append - will append to the end of the file

"w" - Write - will overwrite any existing content

```python
# Open the file "demo.txt" and append content to the file:

my_file = open("demo.txt", "a")
my_file.write("\nNow the file has more content!")

# open and read the file after the appending.
my_file = open("demo.txt", "r")

print(my_file.read())
```

### Overwrite Existing Content
To overwrite the existing content to the file, use the w parameter:


```python
# if we don't have file the (w) parameter create new file automatically.
my_file = open("demo2.txt", "w")       
my_file.write("\nOh noo, I have deleted the content")

# open and read the file after the appending.
my_file = open("demo2.txt", "r")

print(my_file.read())

# Note: the "w" method will overwrite the entire file.
```

```python
# "x" - Create - will create a file, returns an error if the file exists
```

```python
my_file = open("demo4.txt", "x")
my_file.write("The new file with 'x' parameter is created.")

my_file = open("demo4.txt", "r")
print(my_file.read())
```

```python
#If the file already exist, an error will be raised.
f = open("demo.txt", "x")
```

# Delete a File
To delete a file, you must import the OS module, and run its os.remove() function:


```python
import os
os.remove("myfile.txt")
```

### Check if File exist
To avoid getting an error, you might want to check if the file exists before you try to delete it:



```python
import os
if os.path.exists("myfile.txt"):
    os.remove("myfile.txt")
else:
    print("The File name 'myfile.txt' does not exist")
```
