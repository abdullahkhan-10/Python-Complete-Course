# Python OOP
OOP stands for Object-Oriented Programming.

Python is an object-oriented language, allowing you to structure your code using classes and objects for better organization and reusability.

## Classes and Objects
A class defines what an object should look like, and an object is created based on that class.

For example:
Private jet is class, and its objects are Gulfstream Aerospace, Bombardier, Dassault Aviation and Cessna.

When you create an object from a class, it inherits all the variables(Properties) and methods(Behaviors) defined inside that class.


### Create a Class

```python
# To create a class, use the keyword class:
class privateJet:
    # variables(Properties)
    color = "white"

# Now we can use the class named privateJet to create objects: (We can create more than one objects for a class)
gulfstream = privateJet()
dassault = privateJet()

# Now print the property for gulfstream object.
print(gulfstream.color)


# Note: Each object is independent and has its own copy of the class properties.
```

```python
class privateJet:
    # methods(Behaviors)
    def flight_range(self):
        print("The Gulfstream G800 can fly upto 15,186km")

gulfstream = privateJet()

# We can access and print the method in two ways.

# 1
privateJet.flight_range(gulfstream)

# 2, In this we are not passing any arguments, Because in this the object gulfstream is by default pass as argument.
gulfstream.flight_range()
```

#  __init__() Method
All classes have a built-in method called __init__(), which is always executed when the class is being initiated.

The __init__() method is used to assign values to object properties, or to perform operations that are necessary when the object is being created.

```python
class Person:
    def __init__(self, name, age):
        self.my_name = name
        self.my_age = age

# with __init__ method we can set initial values when creating the object.
p1 = Person("Abdullah", 23)           

print(p1.my_name)
print(p1.my_age)

# Note: The __init__() method is called automatically every time the class is being used to create a new object.
```

### Default Values
We can also set default values for parameters in the __init__() method:


```python
class Person:
    def __init__(self, name = "Ali", age = 20):
        self.my_name = name
        self.my_age = age

p1 = Person("Abdullah", 23)
p2 = Person()

print(p1.my_name, p1.my_age)
print(p2.my_name, p2.my_age)
```

### Multiple Parameters
The __init__() method can have as many parameters as we need:

```python
class Person:
    def __init__(self, name, age, city, country):
        self.my_name = name
        self.my_age = age
        self.my_city = city
        self.my_country = country

p1 = Person("Abdullah", 23, "Peshawar", "Pakistan")


print(p1.my_name)
print(p1.my_age)
print(p1.my_city)
print(p1.my_country)
```
