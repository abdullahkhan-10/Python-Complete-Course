# Python OOP
OOP stands for Object-Oriented Programming.

Python is an object-oriented language, allowing you to structure your code using classes and objects for better organization and reusability.

## Classes and Objects
A class defines what an object should look like, and an object is created based on that class.

For example:
Private jet is class, and its objects are Gulfstream Aerospace, Bombardier, Dassault Aviation and Cessna.

When you create an object from a class, it inherits all the variables(Properties) and methods(Behaviors) defined inside that class.


### Create a Class

```python
# To create a class, use the keyword class:
class privateJet:
    # variables(Properties)
    color = "white"

# Now we can use the class named privateJet to create objects: (We can create more than one objects for a class)
gulfstream = privateJet()
dassault = privateJet()

# Now print the property for gulfstream object.
print(gulfstream.color)


# Note: Each object is independent and has its own copy of the class properties.
```

```python
class privateJet:
    # methods(Behaviors)
    def flight_range(self):
        print("The Gulfstream G800 can fly upto 15,186km")

gulfstream = privateJet()

# We can access and print the method in two ways.

# 1
privateJet.flight_range(gulfstream)

# 2, In this we are not passing any arguments, Because in this the object gulfstream is by default pass as argument.
gulfstream.flight_range()
```
