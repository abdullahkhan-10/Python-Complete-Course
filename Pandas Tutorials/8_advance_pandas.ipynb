# Adding Column
We can add new column in two different ways.

### Assigning method
An assigning method the column is added at any index, not specific.

```python
import pandas as pd

df = pd.read_csv("employee_data.csv")

# Increase salary by 10% and assign it to new column "Bonus"
df["Bonus"] = df["salary"] * 0.1

print(df)
```

### insert() Function
Through insert function we can add column at specific index.

```python
# syntax
# df.insert(index, "column_name", some_data)

df = pd.read_csv("employee_data.csv")

df.insert(0, "emp_id", [1001, 1002, 1003, 1004, 1005])
print(df)
```

# Updating Column & Row
We can also update rows and columns in two different ways.

### Assigning method
In assigning method we can update the whole column.

```python
df = pd.read_csv("employee_data.csv")

# increase the salary column by 5%
df["salary"] = df["salary"] * 1.05
print(df)
```

### loc Method
in loc method we can update / set specifi value in specific column.

```python
# Syntax
# df.loc[row_index, "Column_name"] = new_value

df = pd.read_csv("employee_data.csv")

df.loc[2, "salary"] = 57000

print(df)
```

# Removing Column & Row
We will use the drop() and dropna() functions to remove column and row from the dataset.

### drop() Function
The drop() function will remove specific column from the dataframe.

```python
# drop() syntax
# df.drop(columns = ["Column_name"], inplace = True)

df = pd.read_csv("employee_data.csv")

# the drop() function will remove specific column from the dataset.
df.drop(columns = ["number"], inplace = True)

print(df)
```

### dropna() Function
The dropna will remove all rows containing NULL values from the original DataFrame.

```python
# dropna() syntax
# df.dropna(inplace = True)

df = pd.read_csv("employee_data.csv")

df.dropna(inplace = True)

print(df)
```

# Identify Missing Values.
To identify missing value in dataframe we will use isnull() function.

The function return True for Null value and False for non Null value.

```python
df = pd.read_csv("employee_data.csv")

print(df.isnull())
```

```python
# sum() function
# The sum() function tell us about total number of Null value in each row.

df = pd.read_csv("employee_data.csv")

print(df.isnull().sum())
```

# Handling Missing Value
To handle Null / Empty / Missing value we use fillna() function.

Instead of deleting entire rows just because of some empty cells, We can insrt a new value in empty cell.

```python
# Replace NULL values in all empty cells with the number 50

df = pd.read_csv("employee_data.csv")

df.fillna(50, inplace = True)

print(df)
```

```python
# Replace NULL values in specific "salary" columns with the number 45000

df = pd.read_csv("employee_data.csv")

df.fillna({"salary" : 45000}, inplace = True)

print(df)
```

```python
# Calculate the MEAN of column "age" and "salary" and replace any empty values with mean value, within these columns.

df = pd.read_csv("employee_data.csv")

mean_age = df["age"].mean()
mean_salary = df["salary"].mean()

df.fillna({"age" : mean_age}, inplace = True)
df.fillna({"salary" : mean_salary}, inplace = True)

print(df)
```
