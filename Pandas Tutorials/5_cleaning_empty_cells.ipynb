# Data Cleaning
Data cleaning means fixing bad data in your data set.

Bad data could be:

1. Empty cells
2. Data in wrong format
3. Wrong data
4. Duplicates

The data set contains some empty cells ("Date" in row 22, and "Calories" in row 18 and 28).

The data set contains wrong format ("Date" in row 26).

The data set contains wrong data ("Duration" in row 7).

The data set contains duplicates (row 11 and 12).

##  Cleaning Empty Cells
Empty cells can potentially give us a wrong result when you analyze data.


```python
import pandas as pd

df = pd.read_csv("fitness_data.csv")
print(df)
```

### Remove rows
One way to deal with empty cells is to remove rows that contain empty cells.

This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.

```python
# Return a new Data Frame with no empty cells:

df = pd.read_csv("fitness_data.csv")
new_df = df.dropna()

print(new_df)

# Note: By default, the dropna() method returns a new DataFrame, and will not change the original.
```

```python
# Remove all rows with NULL values:

df = pd.read_csv("fitness_data.csv")

# If we want to change the original DataFrame, use the inplace = True argument inside dropna() method.
df.dropna(inplace = True)

print(df)

# Note: Now, the dropna(inplace = True) will NOT return a new DataFrame,
# but it will remove all rows containing NULL values from the original DataFrame.
```

### Replace Empty Values
Another way of dealing with empty cells is to insert a new value instead.

This way you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with a value:

```python
# Replace NULL values with the number 125.

df = pd.read_csv("fitness_data.csv")
df.fillna(125, inplace = True)

print(df)

# NOTE This will replaces all empty cells in the whole Data Frame. 
```

### Replace Only For Specified Columns
To only replace empty values for one column, specify the column name for the DataFrame.


```python
# Replace NULL values in the "Calories" columns with the number 125

df = pd.read_csv("fitness_data.csv")
df.fillna({"Calories" : 125} , inplace = True)

print(df)
```

### Replace Empty Cell Using Mean, Median and Mode
A common way to replace empty cells, is to calculate the mean, median or mode value of the column.

Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column:

```python
# Calculate the MEAN of column "Calories", and replace any empty values with it.

df = pd.read_csv("fitness_data.csv")

# First we have to find the mean value of column "Calories".
mean_value = df["Calories"].mean()

# Now replace the empty value with the mean value.
df.fillna({"Calories" : mean_value}, inplace = True )

print(df)
```

```python
#  Calculate the MEDIAN of column "Calories", and replace any empty values with it.

df = pd.read_csv("fitness_data.csv")

# First we have to find the median value of column "Calories".
median_value = df["Calories"].median()

# Now replace the empty value with the median value.
df.fillna({"Calories" : median_value}, inplace = True)

print(df)
```

```python
#  Calculate the MODE of column "Calories", and replace any empty values with it.

df = pd.read_csv("fitness_data.csv")

# First we have to find the mode value of column "Calories".
mode_value = df["Calories"].mode()[0]

# Now replace the empty value with the mode value.
df.fillna({"Calories" : mode_value}, inplace = True)

print(df)
```
