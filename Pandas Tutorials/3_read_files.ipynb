# Pandas Read CSV
A simple way to store big data sets is to use CSV files (comma separated files).

CSV files contains plain text and is a well know format that can be read by everyone including Pandas.

```python
import pandas as pd

industry_df = pd.read_csv("industry.csv")

print(industry_df)
```

### max_rows
The number of rows returned is defined in Pandas option settings.

We can check our system's maximum rows with the pd.options.display.max_rows statement.

```python
print(pd.options.display.max_rows)

# Note. In my system the number is 60, which means that if the DataFrame contains more than 60 rows,
# the print(industry_df) statement will return only the headers and the first and last 5 rows.
```

# Pandas Read JSON
Big data sets are often stored, or extracted as JSON.

JSON is plain text, but has the format of an object, and is well known in the world of programming, including Pandas.

```python
df = pd.read_json("data.json")

print(df)
```

### Dictionary as JSON
If our JSON code is not in a file, but in a Python Dictionary, we can load it into a DataFrame directly:

```python
data = {
    'Name': ['Khan', 'Salman', 'Saad', 'Ayeza'],
    'Age': [23,17,21,19],
    'Salary': [90000, 70000, 50000, 30000]
}

df = pd.DataFrame(data)
print(df)
```
