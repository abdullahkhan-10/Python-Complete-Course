# Analyzing DataFrames
We have some built in functions in pandas for analyzing large amount of data.

### Head() Method
One of the most used method for getting a quick overview of the DataFrame, is the head() method.

The head() method returns the headers and a specified number of rows, starting from the top.

```python
import pandas as pd

df = pd.read_csv("renewable_energy.csv")

# It will Display first five of data set.
print(df.head())
```

```python
df = pd.read_csv("renewable_energy.csv")

# It will Display first ten rows of data set.
print(df.head(10))
```

### Tail Method
There is also a tail() method for viewing the last rows of the DataFrame.

The tail() method returns the headers and a specified number of rows, starting from the bottom.

```python
df = pd.read_csv("renewable_energy.csv")

# It will Display last ten rows of data set.
print(df.tail(10))
```

### info() Method
The DataFrames object has a method called info(), that gives you more information about the data set.

The information contains the number of columns, column labels, column data types, memory usage, range index, and the number of cells in each column (non-null values).

```python
df = pd.read_csv("renewable_energy.csv")

print(df.info())
```

### describe() Method
The describe() method returns description of the data in the DataFrame.

If the DataFrame contains numerical data, the description contains folowing information for each column,

1. Count
2. Mean
3. std
4. min
5. 25%
6. 50%
7. 75%
8. max

```python
df = pd.read_csv("renewable_energy.csv")

print(df.describe())
```

# Selecting Columns
To select one or more than one column from data set we will use the following syntax.


```python
data = {
    "first_name" : ["Ali", "Jan", "shan", "Ahsan", "Ahmad"],
    "last_name" : ["khan", "Afridi", "Pathan", "Abid", "Danyal"],
    "age" : [23, 25, 27, 19, 16],
    "number" : ["7349282382", "5678568567", "456754675", "7345422382", "1245422345"]
}

df = pd.DataFrame(data)

# Single column return series.
print(df["last_name"])
```

# Filtering Rows
To filter rows from the data set we will use the following syntax.

```python
data = {
    "first_name" : ["Ali", "Jan", "shan", "Ahsan", "Ahmad"],
    "last_name" : ["khan", "Afridi", "Pathan", "Abid", "Danyal"],
    "age" : [23, 25, 27, 19, 16],
    "salary" : [45000, 40000, 60000, 35000, 65000]
}

df = pd.DataFrame(data)

# Display the row which has the highest salary.
high_salary = df[df["salary"] > 45000]
print(high_salary)
```

### and operator (&)

```python
# filter the row which has salary less than 60k and greater than 20 .

data = {
    "first_name" : ["Ali", "Jan", "shan", "Ahsan", "Ahmad"],
    "last_name" : ["khan", "Afridi", "Pathan", "Abid", "Danyal"],
    "age" : [23, 21, 25, 16, 27],
    "salary" : [45000, 40000, 60000, 35000, 65000]
}

df = pd.DataFrame(data)

filtered = df[(df["salary"] < 60000) & (df["age"] > 20)]
print(filtered)
```

### or operator ( | )

```python
# filtere all the rows which fulfill any of the two conditions.
 
data = {
    "first_name" : ["Ali", "Jan", "shan", "Ahsan", "Ahmad"],
    "last_name" : ["khan", "Afridi", "Pathan", "Abid", "Danyal"],
    "age" : [23, 21, 25, 16, 27],
    "salary" : [45000, 40000, 60000, 35000, 65000]
}

df = pd.DataFrame(data)

filtered = df[(df["salary"] > 40000) | (df["age"] < 20)]

print(filtered)
```
