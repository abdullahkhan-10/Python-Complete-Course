# Pandas Series
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

```python
import pandas as pd

arr = [1, 2, 3, 4]

my_series = pd.Series(arr)
print(my_series)
```

### Labels
If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

This label can be used to access a specified value.

```python
arr = [1, 2, 3, 4]

my_series = pd.Series(arr)
print(my_series[2])
```

### Create Labels
With the index argument, you can name your own labels.


```python
arr = [1, 2, 3, 4]

my_series = pd.Series(arr, index = ["x", "y", "z", "a"])
print(my_series)
```

```python
# When you have created labels, you can access an item by referring to the label.
arr = [1, 2, 3, 4]

my_series = pd.Series(arr, index = ["x", "y", "z", "a"])
print(my_series["a"])
```

### Key/Value Objects as Series
We can also use a key/value object, like a dictionary, when creating a Series.

```python
ages = {"ali": 18, "Khan": 20, "Afridi": 23, "jan": 19}

person_ages = pd.Series(ages)
print(person_ages)

# Note: The keys of the dictionary become the labels.
```

# Pandas DataFrame
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

Series is like a column, a DataFrame is the whole table.

```python
employee = {
    "name": ["Khan", "Umar", "Usman", "Ammar"],
    "age": [23, 25, 19, 15],
}

#load data into a DataFrame object:
employee_df = pd.DataFrame(employee)

print(employee_df)
```

### Locate Row
As you can see from the result above, the DataFrame is like a table with rows and columns.

Pandas use the loc attribute to return one or more specified row(s)

```python
# To access the third row of dataframe.
employee = {
    "name": ["Khan", "Umar", "Usman", "Ammar"],
    "age": [23, 25, 19, 15],
}

employee_details = pd.DataFrame(employee)

print(employee_details.loc[2])

# Note: This example returns a Pandas Series.
```

```python
# To access a list of rows/indexes.
employee = {
    "name": ["Khan", "Umar", "Usman", "Ammar"],
    "age": [23, 25, 19, 15],
}

employee_details = pd.DataFrame(employee)

print(employee_details.loc[1 : 3])

# Note: This example returns a Pandas DataFrame.
```

### Named Indexes
With the index argument, you can name your own indexes.


```python
employee = {
    "name": ["Khan", "Umar", "Usman", "Ammar"],
    "age": [23, 25, 19, 15],
}

employee_df = pd.DataFrame(employee, index = ["a", "b", "c", "d"])

print(employee_df)
```

### Locate Named Indexes
Use the named index in the loc attribute to return the specified row(s).


```python
employee = {
    "name": ["Khan", "Umar", "Usman", "Ammar"],
    "age": [23, 25, 19, 15],
}

employee_df = pd.DataFrame(employee, index = ["a", "b", "c", "d"])

print(employee_df.loc["c"])
```
