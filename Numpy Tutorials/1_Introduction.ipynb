# NumPy Introduction
NumPy is a Python library used for working with arrays. NumPy stands for Numerical Python

It also has functions for working in domain of linear algebra, fourier transform, and matrices.

NumPy was created in 2005 by Travis Oliphant. It is an open source project and we can use it freely. It is also used for working with arrays


## Why Use NumPy
In Python we have lists that serve the purpose of arrays, but they are slow to process.

NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.

The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.

```python
# First we have to import numpy
import numpy

# Create numpy array.
my_arr = numpy.array([1, 2, 3, 4, 5])

print(my_arr)
```

### Create a NumPy ndarray Object
NumPy is used to work with arrays. The array object in NumPy is called ndarray.

We can create a NumPy ndarray object by using the array() function.

```python
# NumPy is usually imported under the np alias
import numpy as np

my_arr = np.array([1, 2, 3, 4, 5])

print(my_arr)

# To check the type
print(type(my_arr))
```

```python
# NOTE. To create an ndarray, we can pass a list, tuple or any array-like object into the array() method,
# and it will be converted into an ndarray
```

```python
import numpy as np

# Use a tuple to create a NumPy array
my_arr = np.array((1, 2, 3, 4, 5))

print(my_arr)
```

# Dimensions in Arrays
A specific direction or axis in which we can vary the specification of an array's elements

### 0-D Arrays
0-D arrays, or Scalars, are the elements in an array. Each value in an array is a 0-D array.


```python
import numpy as np

my_arr = np.array(10)

print(my_arr)

# Check Number of Dimensions
print("The Dimension of this array is: ", my_arr.ndim)
```

### 1-D Arrays
An array that has 0-D arrays as its elements is called uni-dimensional or 1-D array.

These are the most common and basic arrays.

```python
import numpy as np

my_arr = np.array([10, 20, 30, 40, 50])

print(my_arr)

# Check Number of Dimensions
print("The Dimension of this array is: ", my_arr.ndim)

# NOTE. NumPy Arrays provides the ndim attribute that returns an integer that tells us how many dimensions the array have.
```

### 2-D Arrays
An array that has 1-D arrays as its elements is called a 2-D array.

These are often used to represent matrix or 2nd order tensors.

```python
import numpy as np

my_arr = np.array([[10, 20, 30], [40, 50, 60]])

print(my_arr)

# Check Number of Dimensions
print("The Dimension of this array is: ", my_arr.ndim)
```

```python
import numpy as np

my_arr = np.array([[[1, 2, 3], [4, 5, 6]], [[10, 15, 20], [25, 30 ,35]]])

print(my_arr)

# Check Number of Dimensions
print("The Dimension of this array is: ", my_arr.ndim)
```

### Higher Dimensional Arrays
An array can have any number of dimensions.

When the array is created, We can define the number of dimensions by using the ndmin argument.

```python
import numpy as np

my_arr = np.array([1, 2, 3, 4], ndmin = 5)

print(my_arr)
print("The Dimensions of this array is: ", my_arr.ndim)
```
