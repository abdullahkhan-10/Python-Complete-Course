# Numpy Initialization Methods
We have some built in methods in numpy that help in initializing numpy array.


### Zeros Array
We will have all the elements 0 in this array.

```python
import numpy as np

zero_arr = np.zeros((2,3))
print(zero_arr)
```

### ones array
We will have all the elements 1 in this array

```python
one_arr = np.ones((1,3))
print(one_arr)
```

```python
one_arr = np.ones((3,4))
print(one_arr)
```

### Full array
We will have all same elements based on given value.

```python
full_arr = np.full((3,2), 5)
print(full_arr)
```

### Identity Matrix
In this array diagonal elements are ones and all other elements are zeros.

```python
id_arr = np.eye(3)
print(id_arr)
```

```python
id_arr = np.eye(4)
print(id_arr)
```

### Empty Array

```python
em_arr = np.empty(2)
print(em_arr)
```

### Evenly spaced array
Elements are evenly spaced based on step value

```python
# Start from 2 and end with 10.

ev_arr = np.arange(2,10,2)
print(ev_arr)
```

```python
# specific number of eqaully spaced values between a range : elements are equally spaced based on step value

ls_arr = np.linspace(1,10,4)
print(ls_arr)
```

### random values array
It will create an array with random value

```python
# Float value array
rf_arr = np.random.rand(3,2)
print(rf_arr)
```

```python
# Integer value array
# The randint() method takes a size parameter where you can specify the shape of an array.

ri_arr = np.random.randint(1,100, size = (3,3))
print(ri_arr)
```
