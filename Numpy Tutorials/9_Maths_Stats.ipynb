# Mathematical Operations on Array
We can do Mathematical Operation on array.

```python
import numpy as np

my_arr = np.array([10, 20, 30, 40])
print(my_arr)
```

```python
my_arr = np.array([10, 20, 30, 40])

print(my_arr + 5)     # add a value in array
print(my_arr - 20)    # Subtract a value in array
print(my_arr * 8)     # Multiply a value in array
print(my_arr / 4)     # Divide a value in array
```

```python
my_arr1 = np.array([1, 4, 8, 9])

print(my_arr1)

print(np.square(my_arr1))       # square of all array elements
print(np.sqrt(my_arr1))         # square root of all array elements
```

## Muliple Arrays
We can do Mathematical operation on Multiple arrays as well.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a)
print(b)
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

### Dot product
Computes the inner product of the two vectors and sum the results

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.dot(a,b))
```

### Transpose Array
Returns an array with axes transposed

```python
my_arr2 = np.array([[1, 2, 3 ],[4, 5, 6]])

print(my_arr2)       # initial shape: 2,3
print(my_arr2.T)     # shape after transpose: 3,2
```

# Statistical Functions
We can do Statistical operation on numpy array.


```python
my_arr3 = np.array([[5, 10, 15, 20],[25, 30, 35, 40]])

print(my_arr3)
```

```python
my_arr3 = np.array([[5, 10, 15, 20],[25, 30, 35, 40]])

print("Statistical Operations")
print(np.sum(my_arr3))
print(np.mean(my_arr3))
print(np.median(my_arr3))
print(np.std(my_arr3))
print(np.min(my_arr3))
print(np.max(my_arr3))
```
