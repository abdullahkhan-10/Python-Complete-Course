# Try Except
The try block lets you test a block of code for errors.

The except block lets you handle the error.



### Exception Handling
When an error occurs, or exception as we call it, Python will normally stop and generate an error message.

These exceptions can be handled using the try statement:

```python
# The try block will generate an exception, because name is not defined:

try:
    print(name)
except:
    print("No name found")

# Since the try block raises an error, the except block will be executed.
```

### Else
We can use the else keyword to define a block of code to be executed if no errors were raised:


```python
# The try block does not generate any erro

try:
    print("Abdullah khan")
except:
    print("Something went wrong")
else:
    print("Nothing went wrong")

# The else statement will execute because no error is found in try statement.
```

### Finally
The finally block, if specified, will be executed regardless if the try block raises an error or not.


```python
# If there is error in try statemnet 
try:
    print(name)
except:
    print("Something went wrong")
finally:
    print("I will execute regardless of error in try statemnet")
```

```python
# If there is no error in try statemnet 
try:
    print("Abdullah khan")
except:
    print("Something went wrong")
finally:
    print("I will execute regardless of error in try statemnet")

# finally statemnnet can be useful to close objects and clean up resources:
```

### Raise an exception
As a Python developer you can choose to throw an exception if a condition occurs.

To throw (or raise) an exception, use the raise keyword.

```python
# Raise an error and stop the program if num is lower than 0:

num = -2

if num < 0:
    raise Exception("Sory, We are not accepting number below zero")

#  if the num is below zero it will throw an error.
```

```python
# Raise a TypeError if my_number is not an integer:

my_number = "abcdefg"

if not type(my_number) is int:
    raise Exception("Only integer are allowed in contact number")
    
```
