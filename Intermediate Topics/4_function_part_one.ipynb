# Python Functions
A function is a block of code which only runs when it is called.
A function can return data as a result.

A function helps avoiding code repetition.

```python
def greeting():
    print("Hi, Abdullah khan")

greeting()
```

### Function Names
A function name must start with a letter or underscore. The name can only contain letters, numbers, and underscores.

Function names are case-sensitive (myFunction and myfunction are different)

```python
def add_number():
    num1 = 10
    num2 = 300
    print(num1 + num2)
add_number()
```

### Return Values
Functions can send data back to the code that called them using the return statement.

When a function reaches a return statement, it stops executing and sends the result back:

```python
def greeting():
    return("Abdullah khan")

# With return value we can store the function in variable.
person = greeting()

print(f"Hi, {person}")

# If a function doesn't have a return statement, it returns None by default.
```

### Why Functions
Imagine you need to convert temperatures from Fahrenheit to Celsius several times in your program.

Without functions, you would have to write the same calculation code repeatedly:

```python
def far_to_cel(farhrenheit):
    return(farhrenheit - 32) * 5 / 9

far_to_cel(104)
```

### pass Statement
Function definitions cannot be empty. If you need to create a function placeholder without any code, use the pass statement:

```python
def my_function():
  pass
```

# Function Arguments
Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, just separate them with a comma.

```python
# A parameter is the variable listed inside the parentheses in the function definition.

def my_name(first_name):          # first_name is a parameter
    print(f"{first_name} Khan")

my_name("Abdullah")     # "Abdullah" is an argument

# An argument is the actual value that is sent to the function when it is called
```

### Number of Arguments
By default, a function must be called with the correct number of arguments.

If your function expects 2 arguments, you must call it with exactly 2 arguments.

```python
def full_name(first_name, last_name):
    print(first_name + " "+ last_name)

full_name("Abdullah", "Khan")
```

### Default Parameter 
We can assign default values to parameters. If the function is called without an argument, it uses the default value:


```python
def greeting(name = "Abdulah"):
    print(f"Hi, {name}")

# The function is called without argument.
greeting()
```

### Keyword Arguments
We can send arguments with the key = value syntax.


```python
def my_bird_name(bird, name):
    print(f"I have a {bird}")
    print(f"my {bird}'s name is {name}")

# With keyword arguments, the order of the arguments does not matter.
my_bird_name(name = "pito" , bird = "Parrot")
```

### Positional Arguments
When you call a function with arguments without using keywords, they are called positional arguments.

Positional arguments must be in the correct order:

```python
def my_bird_name(bird, name):
    print(f"I have a {bird}")
    print(f"my {bird}'s name is {name}")


# The order matters with positional arguments:
my_bird_name("Parrot", "pito" )
```
