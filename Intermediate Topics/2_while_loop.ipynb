# while Loop
With the while loop we can execute a set of statements as long as a condition is true.

```python
i = 1

while i < 5:
    print(i)
    i += 1

# Note: remember we have to increment i, or else the loop will continue forever.
```

```python
name = "Afridi"
i = 0

while i < len(name):   # Condition: it mean that as long as 'i' is less than the string length, loop will execute
    print(name[i])     # Access the character at the current index
    i += 1             # Update: increment the index to move to the next character
```

### break Statement
With the break statement we can stop the loop even if the while condition is true:


```python
# Exit the loop when i is 5:
i = 1

while i < 8:
    print(i)
    if i == 5:
        break
    i += 1
```

### continue Statement
With the continue statement we can stop the current iteration, and continue with the next:


```python
# Continue to the next iteration if i is 4:
i = 0
while i < 8:
    i += 1
    if i == 4:
        continue
    print(i)
```

### else Statement
With the else statement we can run a block of code once when the condition no longer is true:


```python
# Print a message once the condition is false:
i = 1

while i < 6:
    print(i)
    i += 1
else:
    print(" i is no longer less than 6")
```
