# Python Arrays
An array is a special variable, which can hold more than one value at a time.

Python does not have built-in support for Arrays, but Python Lists can be used instead

In this toturial we will learn how to use LISTS as ARRAYS, however, to work with arrays in Python we will have to import a library, like the NumPy library.

```python
birds = ["Eagle", "Parrot", "Falcon", "Pigeon"]

print(birds)
```

### Access the Elements
To access the element of an array we refer to the index number.



```python
birds = ["Eagle", "Parrot", "Falcon", "Pigeon"]

print(birds[0])
print(birds[3])
```

```python
# Modify Value of array
birds = ["Eagle", "Parrot", "Falcon", "Pigeon"]

birds[1] = "Dove"

print(birds)
```

### Length of an Array

```python
birds = ["Eagle", "Parrot", "Falcon", "Pigeon"]

print(len(birds))
```

### Looping Array Elements
We can use the for in loop to loop through all the elements of an array.


```python
birds = ["Eagle", "Parrot", "Falcon", "Pigeon"]

for bird in birds:
    print(bird)
```

### Adding Array Elements
We can use the append() method to add an element to an array.


```python
# append() add element to the end of an array.
birds = ["Eagle", "Parrot", "Falcon", "Pigeon"]

birds.append("Dov")

print(birds)
```

### Removing Array Elements
We can use the pop() method to remove an element from the array.


```python
# Delete the third element of the cars array:
birds = ["Eagle", "Parrot", "Falcon", "Pigeon"]

birds.pop(2)     # pop take index as an argument

print(birds)
```

```python
# We can also use the remove() method to remove an element from the array.
```

```python
birds = ["Eagle", "Parrot", "Falcon", "Pigeon"]

birds.remove("Parrot")         # remove take value as an argument

print(birds)
```
