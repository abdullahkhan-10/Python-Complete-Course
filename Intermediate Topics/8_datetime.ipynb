# Python Datetime
A date in Python is not a data type of its own, but we can import a module named datetime to work with dates as date objects.



```python
# Import the datetime module and display the current date:

import datetime

date_obj = datetime.datetime.now()

print(date_obj)

# The date will contains year, month, day, hour, minute, second, and microsecond.
```

### Creating Date Objects
To create a date, we can use the datetime() class (constructor) of the datetime module.

The datetime() class requires three parameters to create a date: year, month, day.

```python
# We will first import the datetime module
import datetime

date_obj = datetime.datetime(2020, 5, 17)

print(date_obj)

# Note. The datetime() class also takes parameters for time and timezone (hour, minute, second, microsecond, tzone),
# but they are optional, and has a default value of 0
```

### strftime() Method
The datetime object has a method for formatting date objects into readable strings.

The method is called strftime(), and takes one parameter, format, to specify the format of the returned string:

```python
# Display the name of the month:
import datetime

date_obj = datetime.datetime(2002, 7, 15)
print(date_obj.strftime("%B"))     # The strftime() method will print the month in string form.
```

```python
# Return the year and name of weekday:

import datetime
date_obj = datetime.datetime.now()

print(date_obj.year)
print(date_obj.strftime("%A"))
print(date_obj.strftime("%Y"))
```
