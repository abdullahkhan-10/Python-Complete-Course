# Python None
None is a special constant in Python that represents the absence of a value, Its data type is NoneType.

```python
x = None
print(x)
```

```python
x = None
print(type(x))
```

### Comparing to None
To compare a value to None, use the identity operator is or is not



```python
value = None

if value is None:
    print("No value found")
else:
    ("Value is found")
```

### True or False
None evaluates to False in a boolean context.


```python
print(bool(None))
```

### Functions returning None
Functions that do not explicitly return a value, return None by default.


```python
def my_name():
    name = "Abdullah"

x = my_name()

print(x)
```

```python
# With return statement, it will return that value.
def my_name():
    name = "Abdullah"
    return name

x = my_name()

print(x)
```
