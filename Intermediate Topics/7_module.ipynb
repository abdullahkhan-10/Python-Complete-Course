# Python Modules
A module is Consider to be the same as a code library.

A file containing a set of functions you want to include in your application.

### Create a Module
To create a module just save the code you want in a file with the file extension .py:

```python
# Save this code in a file named my_module.py

def greeting(name):
  print(f" Hi, {name}")
```

### Use a Module
Now we can use the module we just created, by using the import statement:

```python
# Import the module named my_module in another file, and access the greeting function:
import my_module

my_module.greeting("Abdullah")

# In jupyter notebook we don't have a proper file system, that's why it will throw an error.
```

### Variables in Module
The module can contain functions, as already described, but also variables of all types (arrays, dictionaries, objects etc):

```python
# Save this code in the file demo.py
person = {
  "name": "Abdullah",
  "age": 23,
  "country": "Pakistan"
}
```

```python
# Import the module named demo in another file, and access the person dictionary:
import demo

my_age = demo.person["age"]
print(my_age)
```

# Math Functions
Python has a set of built-in math functions, including an extensive math module, that allows you to perform mathematical tasks on numbers.


### min() and max()


```python
x = min(5, 10, 25)
y = max(5, 10, 25)

print(x)
print(y)
```

### abs()
The abs() function returns the absolute (positive) value of the specified number:


```python
x = abs(-10.25)

print(x)
```

### pow(x, y)
The pow(x, y) function returns the value of x to the power of y.


```python
#Return the value of 8 to the power of 4 (same as 8 * 8 * 8 * 8):

x = pow(8, 4)

print(x)
```

# Math Module
Python has also a built-in module called math, which extends the list of mathematical functions.

To use it, you must import the math module:

```python
import math

# When We have imported the math module, We can start using methods and constants of the module.
```

### math.sqrt()
The math.sqrt() method returns the square root of a number:


```python
import math

num_square_root = math.sqrt(128)

print(num_square_root)
```

### math.ceil()
The math.ceil() method rounds a number upwards to its nearest integer

```python
import math

num = math.ceil(3.6)


print(num) # returns 4
```

### math.floor()
The math.floor() method rounds a number downwards to its nearest integer

```python
import math

num = math.floor(3.6)


print(num) # returns 3
```

### math.pi
The math.pi constant, returns the value of PI (3.14...):


```python
import math

pi_value = math.pi

print(pi_value)
```
